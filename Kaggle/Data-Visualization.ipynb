# Bar Charts and Heatmaps

## Load the data

```python
# path of the file to read
flight_filepath = "../Kaggle/data/flight_delays.csv"

# read the file into a variable flight_data
flight_data = pd.read_csv(flight_filepath, 
                         index_col = "Month")

flight_data.head()
```

# Scatter Plots
## Load and examine the data

```python
# path of the file to read
insurance_filepath = "../Kaggle/data/insurance.csv"

# read the file into a variable insurance_data
insurance_data = pd.read_csv(insurance_filepath)
```

```python
insurance_data.head()
```

## Scatter plots
To create a simple **scatter plot**, we use the `sns.scatterplot` command and specify the values for:  

- the horizontal x-axis (`x = insurance_data['bmi']`)
- the vertically y-axis (`y = insurance_data['charges']`)

```python
sns.scatterplot(x = insurance_data['bmi'], 
               y = insurance_data["charges"])
```

```python
# regression scatter plot
sns.regplot(x = insurance_data['bmi'], 
           y = insurance_data['charges'], 
           ci = 95)
```

## Color-coded scatter plots

```python
sns.scatterplot(x = insurance_data['bmi'], 
               y = insurance_data["charges"], 
               hue = insurance_data['smoker'])
```

```python
sns.scatterplot(x = insurance_data['bmi'], 
               y = insurance_data["charges"], 
               hue = insurance_data['region'])
```

```python
sns.lmplot(x = "bmi", 
           y = "charges", 
          hue = "smoker",
          data = insurance_data)
```

```python
sns.swarmplot(x = "smoker", 
             y = "charges", 
             data = insurance_data)
```

```python
plt.figure(figsize=(10, 5))

plt.title(label = "categorical plot")

sns.stripplot(x = "smoker", 
             y = "charges", 
             data = insurance_data, 
             dodge = True, 
             hue = "region")

plt.xlabel(xlabel = "smoker")
plt.ylabel(ylabel = "charges")
```

# Distributions
## Load and examine the data

```python
# path of the file to read
iris_filepath = "../Kaggle/data/iris.csv"

# read the file into a variable iris_data
iris_data = pd.read_csv(iris_filepath, 
                       index_col = "Id")

# print the first 5 rows of the data
iris_data.head()
```

## Histograms

```python
sns.distplot(a = iris_data['Petal Length (cm)'], 
            kde = True, 
            bins = 7)
```

## Density plots

The next type of plot is a **kernel density estimate (KDE)** plot. In case you're not familiar with KDE plots, you can think of it as a smoothed histogram.  

To make a KDE plot, we use the `sns.kdeplot`. Setting `shade = True` colors the area below the curve.

```python
sns.kdeplot(x = iris_data['Petal Length (cm)'], 
           shade = True)
```

## 2D KDE plots

We're not restricted to a single column when creating a KDE plot. We can create a **two-dimentional (2D) KDE plot** with the `sns.jointplot`. 

```python
# 2D KDE plot
sns.jointplot(x = iris_data['Petal Length (cm)'], 
             y = iris_data["Sepal Width (cm)"], 
             kind = "kde", 
             shade = True, 
             fill = True)
```

## Color-coded plots

```python
# split the data into three parts based on species
iris_set_filepath = "../Kaggle/data/iris_setosa.csv"
iris_ver_filepath = "../Kaggle/data/iris_versicolor.csv"
iris_vir_filepath = "../Kaggle/data/iris_virginica.csv"

# read the files into variables
iris_set_data = pd.read_csv(iris_set_filepath, index_col="Id")
iris_ver_data = pd.read_csv(iris_ver_filepath, index_col="Id")
iris_vir_data = pd.read_csv(iris_vir_filepath, index_col="Id")

# print the first 5 rows of iris versicolor data
iris_vir_data.head()
```

```python
plt.figure(figsize=(14, 6))

sns.histplot(data = iris_set_data["Petal Length (cm)"], 
             label = "Iris-setosa", 
            kde = False, 
            color = "red")

sns.histplot(data = iris_ver_data["Petal Length (cm)"], 
            label = "Iris-versicolor", 
            kde = False, 
            color = "blue")

sns.histplot(data = iris_vir_data["Petal Length (cm)"], 
            label = "Iris-virginica", 
            kde = False, 
            color = "orange")

plt.title(label = "Histogram of Petal Lengths, By species")

plt.legend()
```

In this case, the legend does not automatically appear on the plot. To force it to show (for any plot), we can always use `plt.legend()`. 

```python
sns.kdeplot(data = iris_set_data["Petal Length (cm)"], 
           label = "Iris-setosa", 
           shade = True)

sns.kdeplot(data = iris_ver_data["Petal Length (cm)"], 
           label = "Iris-setosa", 
           shade = True)

sns.kdeplot(data = iris_vir_data["Petal Length (cm)"], 
           label = "Iris-setosa", 
           shade = True)

# add title
plt.title(label = "Distribution of Petal Lengths, by Species")

plt.legend()
```

# Choosing Plot Types and Custom Styles

Since it's not always easy to decide how to best tell the story behind your data, we've broken the chart types into three broad categories to help with this.  

- **Trends** - A trend is defined as a pattern of change.  
    + `sns.lineplot` - **Line charts** are best to show trends over a periods of time, and multiple lines can be used to show trends in more than one group.  
- **Relationship** - There are many different chart types that you can use to understand relationships between variables in your data.  
    + `sns.barplot` - **Bar charts** are useful for comparing quantities corresponding to different groups.  
    + `sns.heatmap` - **Heatmaps** can be used to find color-coded patterns in tables of numbers.  
    + `sns.scatterplot` - **Scatter plot** show the relationship between two continuous variables; if color-coded, we can also show the relationship with a third **categorical variable**.  
    + `sns.regplot` - Including a **regresion line** in the scatter plot makes it easier to see any linear relationship between two variables.  
    + `sns.lmplot` - This command is usedul for drawing multiple regression lines, if the scatter plot contains multiple, color-coded groups.  
    + `sns.swarmplot` - **Categorical scatter plots** show the relationship between a continous variable and a categorical variable.  
- **Distribution** - We visualize distributions to show the possible values that we can expect to see n a variable, along with how likely they are.  
    + `sns.distplot` - **Histograms** show the distribution of a single numerical variable.  
    + `sns.kdeplot` - **KDE plots** (or **2D KDE plots**) show an estimated, smooth distribution of a single numerical variable (or two numerical variables).  
    + `sns.jointplot` - This command is useful for simultaneously displaying 2D KDE plot with the corresponding KDE plots for each individual variable.
