## Combining and Merging Datasets

- pandas.merge: Connect rows in DataFrames based on one or more keys.
    + inner merge
    + left merge
    + right ,merge
    + outer merge
- pandas.concat: Concatenate or "stack" objects together along an axis.
- combine_first: Splice together overlapping data to fill in missing values in one object with values from another.

### Database-Style Dataframe Joins

```python
df1 = pd.DataFrame({"key": ["b", "b", "a", "c", "a", "a", "b"], 
                   "data1": pd.Series(range(7), dtype = "Int64")})
df1
```

```python
df2 = pd.DataFrame({"key": ["a", "b", "d"], 
                   "data2": pd.Series(range(3), dtype = "Int64")})
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on = "key")
```

```python
df3 = pd.DataFrame({"lkey": ["b", "b", "a", "c", "a", "a", "b"], 
                  "data1": pd.Series(range(7), dtype = "Int64")})
df3
```

```python
df4 = pd.DataFrame({"rkey": ["a", "b", "d"], 
                   "data2": pd.Series(range(3), dtype = "Int64")})
df4
```

```python
pd.merge(df3, df4, left_on="lkey", right_on="rkey")
```

```python
pd.merge(df1, df2, how = "outer")
```

```python
df1 = pd.DataFrame({"key": ["b", "b", "a", "c", "a", "b"], 
                   "data1": pd.Series(range(6), dtype = "Int64")})
df1
```

```python
df2 = pd.DataFrame({"key": ["a", "b", "a", "b", "d"], 
                   "data2": pd.Series(range(5), dtype = "Int64")})
df2
```

```python
pd.merge(df1, df2, on = "key", how = "left")
```

```python
pd.merge(df1, df2, on = "key", how = "inner")
```

```python
left = pd.DataFrame({"key1": ["foo", "foo", "bar"], 
                    "key2": ["one", "two", "one"], 
                    "lval": pd.Series([1, 2, 3], dtype = "Int64")})
left
```

```python
right = pd.DataFrame({"key1": ["foo", "foo", "bar", "bar"], 
                     "key2": ["one", "one", "one", "two"], 
                     "rval": pd.Series([4, 5, 6, 7], dtype = "Int64")})
right
```

```python
pd.merge(left, right, on = ["key1", "key2"], how = "outer")
```

```python
pd.merge(left, right, on = "key1", suffixes=("_left", "_right"))
```

### Merge on Index

```python
left1 = pd.DataFrame({"key": ["a", "b", "a", "a", "b", "c"], 
                     "value": pd.Series(range(6), dtype = "Int64")})
left1
```

```python
right1 = pd.DataFrame({"group_val": [3.5, 7]}, 
                     index = ["a", "b"])
right1
```

```python
pd.merge(left1, right1, left_on="key", right_index=True)
```

```python
pd.merge(left1, right1, left_on = "key", right_index = True, how = "outer")
```

```python
left1.join(right1, on = "key")
```

```python
left1
```

```python
right1
```

```python
another = pd.DataFrame([[7, 8], [9, 10], [11, 12], [16, 17]], 
                      index = ["a", "c", "e", "f"], 
                      columns = ["New York", "Oregon"])
another
```

```python
left1 = left1.set_index("key")
```

```python
left1.join([right1, another])
```

### Combining Data with Overlap

```python
a = pd.Series([np.nan, 2.5, 0.0, 3.5, 4.5, np.nan], 
             index = ["f", "e", "d", "c", "b", "a"])
a
```

```python
b = pd.Series([0, np.nan, 2, np.nan, np.nan, 5], 
             index = ['a', "b", "c", "d", "e", "f"])
b
```

```python
# like ifelse in R
np.where(pd.isna(a), b, a)
```

```python
a.combine_first(b)
```

```python
df1 = pd.DataFrame({"a": [1, np.nan, 5, np.nan], 
                   "b": [np.nan, 2, np.nan, 6], 
                   "c": range(2, 18, 4)})
df1
```

```python
df2 = pd.DataFrame({"a": [5, 4, np.nan, 3, 7], 
                   "b": [np.nan, 3, 4, 6, 8]})
df2
```

```python
df1.combine_first(df2)
```

# Reshaping and Pivoting

```python
data = pd.DataFrame(np.arange(6).reshape((2, 3)), 
                   index = pd.Index(["Ohio", "Colorado"], name = "state"), 
                   columns=pd.Index(["one", "two", "three"], name = "number"))
data
```

```python
result = data.stack()
result
```

```python
result.unstack()
```

```python
result.unstack(level=0)
```

```python
result.unstack(level = "state")
```

```python
s1 = pd.Series([0, 1, 2, 3], index = ["a", "b", "c", "d"], dtype = "Int64")
s2 = pd.Series([4, 5, 6], index = ["c", "d", "e"], dtype = "Int64")
data2 = pd.concat([s1, s2], keys = ["one", "two"])
data2
```

```python
data2.unstack(level=1)
```

```python
data2.unstack(level = 0)
```

```python
data2.unstack().stack(dropna = False)
```

## Pivoting "Long" to "Wide" Format

```python
data = pd.read_csv("../../github/pydata-book/examples/macrodata.csv")
data = data.loc[:, ["year", "quarter", "realgdp", "infl", "unemp"]]
data.head()
```

```python
periods = pd.PeriodIndex(year = data.pop("year"), 
                        quarter = data.pop("quarter"), 
                        name = "date")
periods
```

```python
data.index = periods.to_timestamp("D")
data.head
```

```python
data = data.reindex(columns=["realgdp", "infl", "unemp"])
data.columns.name = "item"
data
```

```python
long_data = data.stack().reset_index().rename(columns={0:"value"})
```

```python
long_data
```

```python
pivoted = long_data.pivot(index = "date", 
                         columns="item", 
                         values="value")
pivoted.head()
```

## Pivoting "Wide" to "Long" Format

```python
df = pd.DataFrame({"key": ["foo", "bar", "baz"], 
                  "A": [1, 2, 3], 
                  "B": [4, 5, 6], 
                  "C": [7, 8, 9]})
df
```

```python
melted = pd.melt(df, id_vars = "key")
melted
```

```python
melted.pivot(index = "key", columns="variable", values = "value")
```

```python
pd.melt(df, id_vars="key", value_vars=["A", "B"])
```
