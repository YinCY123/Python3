## Ticks, Labels, and Legends

```python
fig, ax = plt.subplots()
ax.plot(np.random.standard_normal(1000).cumsum());
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(["one", "two", "three", "four", "five"], 
                           rotation = 30, fontsize = 8)

ax.set_xlabel("Stage")
ax.set_title("My first matplotlib plot")

# ax.set(title = "My first matplotlib plot", xlabel = "Stages")
# same as the above two lines
```

```python
fig, ax = plt.subplots(2, 3)

ax[0, 0].plot(np.random.randn(1000).cumsum(), 
        color = "black", label = "one")
ax[0, 0].plot(np.random.randn(1000).cumsum(), 
        color = "black", linestyle = "dashed", 
        label = "two")

ax[0, 0].plot(np.random.randn(1000).cumsum(), 
        color = "black", linestyle = "dotted", 
       label = "three")
ax[0,0].legend()
```

```python
from datetime import datetime
```

```python
fig, ax = plt.subplots()
data = pd.read_csv("../../github/pydata-book/examples/spx.csv", 
                  index_col=0, parse_dates=True)
data.head()
spx = data["SPX"]
spx

spx.plot(ax = ax, color = "black");

crisis_data = [(datetime(2007, 10, 11), "Peak of bull market"), 
              (datetime(2008, 3, 12), "Bear Stearns Fails"), 
              (datetime(2008, 9, 15), "Lehman Bankruptcy")]

for date, label in crisis_data:
    ax.annotate(label, xy = (date, spx.asof(date) + 75), 
               xytext = (date, spx.asof(date) + 200), 
               arrowprops = dict(facecolor = "black", headwidth = 4, 
                                 width = 2, headlength = 4),
               horizontalalignment = "left", verticalalignment = "top");
    
ax.set_xlim(["1/1/2007", "1/1/2011"]);
ax.set_ylim([600, 1800]);

ax.set_title("Important dates in the 2008 - 2009 finical crisis");
```

```python
fig, ax = plt.subplots()
rect = plt.Rectangle(xy = (0.2, 0.75), width = 0.4, height = 0.15, 
                    color = "black", alpha = 0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color = "blue", alpha = 0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]], 
                  color = "green", alpha = 0.5)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon);
fig.savefig("plots.tiff", dpi = 300)
```

```python
df = pd.DataFrame(np.random.uniform(size = (6, 4)), 
                 index = ["one", "two", "three", "four", "five", "six"], 
                 columns = pd.Index(["A", "B", "C", "D"], name = "Genus"))
df
```

```python
df.plot.bar(color = {"A":"red", "B": "green", "C": "blue", "D": "purple"}, 
           grid = True, 
           rot = 0, 
           fontsize = 20, 
           position = 0.5, 
           yerr = False, 
           ylim = (0, 3.5), 
            xlim = (0, 7),
           stacked = True)
```

```python
df.plot(kind = "bar", stacked = True,
        rot = 0, fontsize = 15, 
       title = "stacked bar plot", 
       legend = True, 
       color = {"A":"red", "B": "green", "C": "blue", "D": "purple"});
```

```python
tips = pd.read_csv("../../github/pydata-book/examples/tips.csv")
tips.head()
```

```python
party_counts = pd.crosstab(tips["day"], tips["size"])
party_counts
```

```python
# reorder rows
party_counts = party_counts.reindex(index = ["Thur", "Fri", "Sat", "Sun"])
party_counts
```

```python
party_counts = party_counts.iloc[:, 2:5]
party_counts
```

```python
party_pcts = party_counts.div(party_counts.sum(axis = 1), 
                             axis = "index")
party_pcts
```

```python
party_counts.plot(kind = "bar", 
                  stacked = True, 
                  rot = 0, 
                  xlabel = "", 
                  fontsize = 15);
```

```python
party_counts.plot.bar(stacked = True, rot = 0, xlabel = "");
```

## Scatter or Points Plots

```python
macro = pd.read_csv("../../github/pydata-book/examples/macrodata.csv")
data = macro[["cpi", "m1", "tbilrate", "unemp"]]
data.head()
trans_data = np.log(data).diff().dropna()
trans_data.tail()
```

```python
ax = sns.regplot(x = "m1", y = "unemp", data = trans_data)
#ax.title("Changes in log(m1) versus log(unemp)")
```

```python
sns.scatterplot(data = trans_data, x = "m1", y = "unemp");
```

```python
sns.pairplot(data=trans_data, diag_kind="kde", plot_kws={"alpha": 0.2})
```

```python
sns.catplot(x = "day", y = "tip_pct", hue = "time", col = "smoker", 
           kind = "bar", data = tips[tips.tip_pct < 1])
```

```python
sns.catplot(x = "day", y = "tip_pct", 
            row = "time", col = "smoker", 
            kind = "bar", data = tips[tips["tip_pct"] < 1], 
            color = "gray");
```

```python
sns.catplot(x = "tip_pct", y = "day", 
           kind = "box", 
           data = tips[tips.tip_pct < 0.5], 
           color = "gray");
```
